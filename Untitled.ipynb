```python
import io
import pandas as pd
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

locations_url = "https://rowermevo.pl/maps/locations.txt?key=jF5puHQe3CqssPZh"
```

```python
def get_data():
    http = urllib3.PoolManager()
    resp = http.request('GET', locations_url)
    return resp.data
```

```python
df = pd.read_csv(io.BytesIO(get_data()), index_col="NUMER STACJI")
print("Średnio rowerów", df["DOSTĘPNE ROWERY"].mean())
print("Dostepnych rowerów", df["DOSTĘPNE ROWERY"].sum())
```

```python
from time import sleep
while True:
    df = pd.read_csv(io.BytesIO(get_data()), index_col="NUMER STACJI")
    print("Średnio rowerów: %.2f " % df["DOSTĘPNE ROWERY"].mean(), end="")
    print("Dostepnych rowerów", df["DOSTĘPNE ROWERY"].sum())
    sleep(5)
```
